```python

import optuna
from optuna.trial import TrialState
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import pandas as pd
import numpy as np

path = "../studies/split_optimizer.db"
name = "OptunaOptimization#011"

study = optuna.load_study(  study_name=name,
                            storage=f"sqlite:///{path}"
)

```

```python
def rgb_to_rgba(rgb_value, alpha):
    """
    Adds the alpha channel to an RGB Value and returns it as an RGBA Value
    :param rgb_value: Input RGB Value
    :param alpha: Alpha Value to add  in range [0,1]
    :return: RGBA Value
    """
    return f"rgba{rgb_value[3:-1]}, {alpha})"
```

```python

qLearningRate = []
learningRate = []

objective_distance = {'obj_value':[], 'distance':[]}

values = {}

fig = go.Figure()
i = 0
for trial in study.trials:
    if trial.state == TrialState.PRUNED or trial.state == TrialState.FAIL:
        continue
    
    if trial.params['optimizer'] not in values:
        values[trial.params['optimizer']] = []
        
    epochs = len(trial.intermediate_values)
    fig.add_trace(go.Scatter(x=[i for i in range(epochs)], y=list(trial.intermediate_values.values()), line=dict(color=rgb_to_rgba(px.colors.qualitative.Dark2[0], 0.9) if trial.params['optimizer'] == 'combined' else rgb_to_rgba(px.colors.qualitative.Dark2[1], 0.15))))



fig.update_layout(
    yaxis=dict(
                showgrid=False,
                title='Accuracy'
            ),
            xaxis=dict(
                showgrid=False,
                title='Epochs'
            ),
            template="simple_white",
)

# fig.add_trace(go.Scatter(x=np.linspace(1,epochs), y=values['combined']))

fig.show()
fig.write_image('export.png')
```
